```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('cat.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axes().get_xaxis().set_visible(False) # 隐藏x坐标轴
plt.axes().get_yaxis().set_visible(False) # 隐藏y坐标轴
plt.imshow(img)
```

```python
# 上采样之unpooling
def unpooling(src, scale):
  src_h, src_w, channel = src.shape
  dst_h = src_h * scale
  dst_w = src_w * scale
  dst = np.zeros((dst_h, dst_w, channel), dtype=src.dtype)
  dst[::scale, ::scale] = src
  return dst

img_u = unpooling(img, 2)
plt.axes().get_xaxis().set_visible(False) # 隐藏x坐标轴
plt.axes().get_yaxis().set_visible(False) # 隐藏y坐标轴
plt.imshow(img_u)
```
